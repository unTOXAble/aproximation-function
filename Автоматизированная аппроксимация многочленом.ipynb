```python
import numpy as np
from scipy import linalg
from matplotlib import pylab as plt
```

```python
f = lambda x:np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
x = np.linspace(1, 16)
plt.plot(x, f(x))
plt.show()
```

```python
xn = [1, 8, 15, 11, 6, 9]
```

```python
yn = []
[yn.append(f(xn[i])) for i in range(len(xn))]
# print(yn)
```

```python
An = []
bn = []

for i in yn:
    yi = []
    yi.append(i)
    bn.append(yi)


for i in xn:
    line = []
    for n in range(len(xn)):
        line.append(i**n)
    An.append(line)

    
An = np.array(An)
bn = np.array(bn)
# print(An)
# print(bn)

wn = linalg.solve(An, bn)
# print(wn)   
```

```python
n = 0
p =''
x = np.linspace(1, 16)
for i in wn:
    p = p  + f"+({float(i)}*x**{n}) "
    n +=1 

func = p
def fn(x):
    return eval(func)

# print(fn(4))
# print(func)
# print(type(func))
# print(type(l))
```

```python
plt.plot(x, f(x), label = 'f(x)')
plt.plot(x, fn(x), label = 'fn(x)')
plt.legend()
plt.show()
```
